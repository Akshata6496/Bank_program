```python
#Guessing game
import random
number=random.randint(1,20)
guess=0
count=0
while guess!=number:
    guess=input("guess the number: ")
    
    if guess=="exit":
        break
    guess=int(guess)
    count=count+1
    
    if guess < number:
        print("you guessed too low")
        
    elif guess > number:
        print("You guessed too high")
        
    else:
        print("you guessed the right number that is",number,"in",count, "tries")
        
    
```
